```python
#Gerekli kütüphanelerin import edilmesi
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from google.colab import files
from skimage.transform import resize
import cv2
import glob
import os
import numpy as np
count = -1
count2 = -1
#Görüntüleri ve bu görüntülere ait etiketlerin tutulacağı listeler
goruntu = []
etiket = []
```

```python
#Görüntüler bulunduğu klasör içerisinden alınır
for main_file in glob.glob("/mydrive/PetImages2"):
    print("yes")
    for category_file in glob.glob(os.path.join(main_file, "*")):
        print("ok")
        #Count ile girilen klasördeki görüntüler için bir değer tutulur.. Bu değer kategoriyi, yani sınıfın değeri = verinin etiket olur.
        count+=1
        for img_path in glob.glob(os.path.join(category_file, "*.jpg")):
            count2 += 1
            #Görüntü okuma
            img = cv2.imread(img_path)
            #NoneType görüntüleri almamak için kullanılan şart
            if img is not None:
              #OpenCV ile görüntüleri griye çevirme
              grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
              #Görüntüleri yeniden boyutlandırma
              resized_img = resize(grayImg, (128,64))
              #Görüntü ve etiketleri listelere ekleme
              goruntu.append(resized_img)
              etiket.append(count)
```

```python
print(count2)
count3 = 0
for i in etiket:
  count3+=1
print(count3)
count4 = 0
for i in goruntu:
  count4+=1
  #print(i)
print(count4)
```

```python
#Görüntü verisinin matrisini tek boyuta indirmek için Scale Normalization yöntemi
scaler = StandardScaler()
train_Data = scaler.fit_transform([i.flatten() for i in goruntu])
```

```python
count4 = 0
for i in train_Data:
  count4+=1
print(count4)
```

```python
#Veri ve etiketlerini X ve y olarak başka bir listeye atama
X = train_Data
y = etiket
```

```python
#Veriyi train ve test olarak ayırma (%20'si test olacak şekilde)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape)
print(len(y_train))
print(X_test.shape)
print(len(y_test))
```

```python
#Train verisini tekrardan train ve val olarak ayırma (%25'i val olacak şekilde)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
print(X_train.shape)
print(len(y_train))
print(X_val.shape)
print(len(y_val))
```
